## Weather Data Analysis:

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_excel("Weather_data.xlsx") 
print("Original Weather Data:")
print(df)
df_filled = df.fillna(25)  
print("\nFilled NaN values:")
print(df_filled)
average_temp = df['temperaure'].mean()
df['average'] = average_temp
print("\nSelected columns with average:")
print(df[['Date', 'temperaure', 'humidity', 'wind speed', 'rainfall', 'average']].head())
plt.figure(figsize=(10, 5))  
df['temperaure'].plot(kind='bar')  
plt.title('Temperature Data')
plt.xlabel('Index')
plt.ylabel('Temperature')
plt.show()
```

## time series

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose

dates = pd.date_range(start='2023-01-01', end='2023-12-31', freq='D')
np.random.seed(42)
base_temp = 20 + 10 * np.sin(2 * np.pi * np.arange(365)/365)
daily_variation = np.random.normal(0, 3, 365)
temperatures = np.round(base_temp + daily_variation, 1)

temp_data = pd.DataFrame({
    'Date': dates,
    'Temperature': temperatures
})
temp_data.set_index('Date', inplace=True)

print("Basic Statistics:")
print(temp_data.describe())

print("\nHottest Day:")
print(temp_data.idxmax())

print("\nColdest Day:")
print(temp_data.idxmin())

monthly_stats = temp_data.resample('ME').agg(['mean', 'max', 'min'])
print("\nMonthly Statistics:")
print(monthly_stats)

plt.figure(figsize=(12, 6))
plt.plot(temp_data.index, temp_data['Temperature'], label='Daily Temperature')
plt.plot(temp_data.resample('W').mean(), label='Weekly Average', linewidth=2)
plt.title('Daily Temperatures Over a Year')
plt.xlabel('Date')
plt.ylabel('Temperature (C)')
plt.legend()
plt.grid(True)
plt.show()

temp_data['7D_Avg'] = temp_data['Temperature'].rolling(window=7).mean()

result = seasonal_decompose(temp_data['Temperature'], model='additive', period=30)
result.plot()
plt.show()
```

## Daily stock prices

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)
trading_days = 252
start_price = 100

daily_returns = np.random.normal(0.0005, 0.02, trading_days)
price_series = start_price * (1 + daily_returns).cumprod()

dates = pd.date_range('2023-01-01', periods=trading_days, freq='B')
stock_data = pd.DataFrame({'Price': price_series}, index=dates)

stock_data['Daily_Return'] = stock_data['Price'].pct_change()
stock_data['20D_MA'] = stock_data['Price'].rolling(window=20).mean()
stock_data['50D_MA'] = stock_data['Price'].rolling(window=50).mean()

print("Stock Price Statistics:")
print(f"Final Price: {stock_data['Price'].iloc[-1]:.2f}")
print(f"Annual Return: {stock_data['Price'].iloc[-1]/start_price-1:.2%}")
print(f"Max Price: {stock_data['Price'].max():.2f}")
print(f"Min Price: {stock_data['Price'].min():.2f}")
print("\nDaily Returns Statistics:")
print(stock_data['Daily_Return'].describe())

plt.figure(figsize=(12, 6))
plt.plot(stock_data['Price'], label='Daily Price', alpha=0.7)
plt.plot(stock_data['20D_MA'], label='20-Day MA', color='orange')
plt.plot(stock_data['50D_MA'], label='50-Day MA', color='red')
plt.title('Simulated Stock Price with Moving Averages')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.grid(True)

plt.figure(figsize=(12, 4))
plt.hist(stock_data['Daily_Return'].dropna(), bins=50, alpha=0.7)
plt.title('Distribution of Daily Returns')
plt.xlabel('Daily Return')
plt.ylabel('Frequency')
plt.grid(True)

plt.show()
```
